```python
from sklearn import datasets
import numpy as np
import pandas as pd
import collections
import seaborn as sn
import matplotlib.pyplot as plt
```

## Load Dataset

```python
iris_dataset = datasets.load_iris()
```

```python
x = iris_dataset.data
y = iris_dataset.target
```

## K-Nearest Neighbor Algorithm

```python
def knn(test, train, target, K):
    num_train = train.shape[0]
    dist = np.sum((train - np.tile(test, (num_train, 1)))**2, axis=1)
    idx = sorted(range(len(dist)), key=lambda i: dist[i])[:K]
    return collections.Counter(target[idx]).most_common(1)[0][0]
```

## Test

### 1-NN

```python
confusion_matrix = np.zeros((3, 3))
```

```python
for i in range(len(x)):
    idx = np.zeros((150))
    idx[i] = 1
    
    x_train = x[idx==0]
    y_train = y[idx==0]
    
    x_test = x[idx==1]
    true_ = y[idx==1]
    
    pred_ = knn(x_test, x_train, y_train, 1)
    confusion_matrix[pred_][true_] += 1
```

```python
print(confusion_matrix)
```

```python
df = pd.DataFrame(confusion_matrix, index=range(3), columns=range(3))
```

```python
sn.heatmap(df, annot=True, annot_kws={"size":15})
plt.show()
```
