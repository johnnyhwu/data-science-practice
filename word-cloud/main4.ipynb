```python
import jieba
```

```python
with open("WordCloud3.txt",'r')as f:
    content = f.read()
```

先讀取文字檔

```python
tags_list = jieba.cut(content)

with open("WorldCloud3 Output.txt",'w') as g:
    for tags in tags_list:
        g.write("%s\\" %(tags))
```

將原始文字檔【分詞】，並且加上【\】寫入新檔案

```python
from wordcloud import WordCloud

with open("WorldCloud3 Output.txt",'r')as h:
    content = h.read()
    
font = "NotoSansMonoCJKtc-Regular.otf"

wc = WordCloud(max_font_size=40,font_path=font).generate(content)
```

製作【文字雲】

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(40,40))

plt.axis("off")

plt.imshow(wc,interpolation="bilinear")
```
