```python
from wordcloud import WordCloud
```

【WordCloud】Module可以產生文字雲

```python
with open("WordCloud2.txt","r") as f:
    content = f.read()

font = "NotoSansMonoCJKtc-Regular.otf"
```

【font】:代表【中文字典】

```python
wc = WordCloud(max_font_size=40,font_path=font).generate(content)
```

```python
import matplotlib.pyplot as plt
```

利用【matplotlib.pyplot】module呈現文字雲

```python
plt.figure(figsize=(15,15))

plt.axis("off")

plt.imshow(wc,interpolation="bilinear")
```
