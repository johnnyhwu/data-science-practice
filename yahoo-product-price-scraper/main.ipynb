```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
Yahoo_URL = "https://tw.bid.yahoo.com/tw/%E6%99%BA%E6%85%A7%E7%A9%BF%E6%88%B4%E8%A3%9D%E7%BD%AE-8704571-category.html?.r=1545468098&hpp=hp_category_8704571"

Yahoo_O_HTML = requests.get(Yahoo_URL)

Yahoo_M_HTML = BeautifulSoup(Yahoo_O_HTML.text,"lxml")
```

```python
name = []
price = []
sell = []
```

```python
namess = Yahoo_M_HTML.select(".BaseGridItem__title___2HWui")

for names in namess:
    try:
        name.append(names.text)
    except:
        name.append(np.nan)
```

```python
pricess = Yahoo_M_HTML.select("em")

for prices in pricess:
    try:
        price.append(prices.text)
    except:
        price.append(np.nan)
```

```python
sellss = Yahoo_M_HTML.select(".BaseGridItem__price___31jkj")

for sells in sellss:
    if("售出" in sells.text):
        spl = sells.text.split('/')
        result = spl[1][4:-2]
        sell.append(float(result))
    else:
        sell.append(np.nan)
```

```python
product_dict = {"Name":name,
               "Price":price,
               "Sell":sell}
product_df = pd.DataFrame(product_dict)

product_df
```

```python
for index,i in enumerate(price):
    if("," in i[1:]):
        lis = i[1:].split(",")
        a = lis[0]+lis[1]
        price[index]=float(a)
    else:
        price[index]=float(i[1:])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(x=range(len(product_df)),height=price,color="green")

plt.xlabel("Product")
plt.ylabel("Price")
plt.title("The Price of Product")

plt.show()
```
